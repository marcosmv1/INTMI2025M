##Jogo Da Adivinhação

```python
import random as rd
```

```python
aleatorio = rd.randint(0,100)

print("==============================================================")
print("                JOGO DE ADIVINHAÇÃO                        ")
print("===============================================================")
print("      Tente adivinhar o número que estou pensando!"         )
print("==============================================================")

tentativa = 5

while (tentativa > 0):
  numero = int(input("Digite um numero: "))
  if (numero == aleatorio):
    print("Parabéns! Você Acerotu!")
    break
  elif (numero > aleatorio):
    print("Digite um numero menor:")
  else:
    print("Digite um numero maior:")
    tentativa -= 1
    if (tentativa == 0):
      print("Você não adivinhou!")
```

```python
# =========================
#   CALCULADORA INTERATIVA
# =========================

while True:
    print("\n=========================")
    print("   CALCULADORA INTERATIVA")
    print("=========================")
    print("Digite o número da opção desejada:")
    print("1 - Múltiplos de 5 em um Intervalo")
    print("2 - Contagem de Números Pares em um Intervalo")
    print("3 - Verificador de Número Primo")
    print("4 - Conversor de Temperatura (Cº para Fº)")
    print("0 - Sair do Programa")

    opcao = input("Opção escolhida: ")

    # ============================
    #        OPÇÃO 0 – SAIR
    # ============================
    if opcao == "0":
        print("Saindo do programa... Até a próxima!")
        break

    # ============================
    #   OPÇÃO 1 – Múltiplos de 5
    # ============================
    elif opcao == "1":
        inicio = int(input("Digite o número inicial: "))
        fim = int(input("Digite o número final: "))

        print(f"\nMúltiplos de 5 entre {inicio} e {fim}:")
        for n in range(inicio, fim + 1):
            if n % 5 == 0:
                print(n)

    # ===================================
    #   OPÇÃO 2 – Contagem de pares
    # ===================================
    elif opcao == "2":
        inicio = int(input("Digite o número inicial: "))
        fim = int(input("Digite o número final: "))

        contagem_pares = 0

        for n in range(inicio, fim + 1):
            if n % 2 == 0:
                contagem_pares += 1

        print(f"\nQuantidade de números pares no intervalo: {contagem_pares}")

    # ===================================
    #   OPÇÃO 3 – Número Primo
    # ===================================
    elif opcao == "3":
        num = int(input("Digite um número inteiro positivo: "))

        if num <= 1:
            print("O número não é primo.")
        else:
            eh_primo = True
            for i in range(2, num):
                if num % i == 0:
                    eh_primo = False
                    break

            if eh_primo:
                print(f"O número {num} é primo!")
            else:
                print(f"O número {num} NÃO é primo.")

    # ===================================
    #   OPÇÃO 4 – Conversão C → F
    # ===================================
    elif opcao == "4":
        c = float(input("Digite a temperatura em Celsius: "))
        f = c * 1.8 + 32
        print(f"{c}°C equivalem a {f:.2f}°F")

    # ===================================
    #   OPÇÃO INVÁLIDA
    # ===================================
    else:
        print("Opção inválida! Tente novamente.")
```
